## KS test to compare 2 samples

```python
# scipy.stats.ks_2samp takes parameters a, b: two arrays of sample observations assumed to be 
# drawn from a continuous distribution, sample sizes can be different

# scipy.stats.ks_2samp returns (D, p value), KS statistic and two-tailed p-value 
ks_duration = scipy.stats.ks_2samp(cb_df.day_duration, cb_df.night_duration)
print(ks_duration)
```

The ks test returned a p-value of 0.0. Since the p-value is less than the the significance level I set ($\alpha = 0.05$) we reject the null hypothesis that there is no statistical difference in the trip duration of day vs. night trips.

```python
# Subsample the data 
day_trips = cb_df_testing['day_duration']
night_trips = cb_df_testing['night_duration']

# Drop NA values 
day_trips.dropna()
night_trips.dropna()

# Find out how what size to make the subsampled datasets
print(len(day_trips) / 200)
print(len(night_trips) / 200)
```

```python
# Set a random seed
np.random.seed(1964)

# Create subsets of size 1500 each for day and night duration 
day_sub = numpy.random.choice(day_trips, 1500, replace=False)
night_sub = numpy.random.choice(night_trips, 1500, replace=False)

# Sort subsets from low to high 
day_sub = np.sort(day_sub)
night_sub = np.sort(night_sub)
```

## Pearson's test for correlation

**H0: there is no correlation between day trip duration and night trip duration**

Signficance level: $\alpha = 0.05$

```python
pearsonr_duration = scipy.stats.pearsonr(day_sub, night_sub)
print(pearsonr_duration)
```

The Pearson R test returned an extremely small p-value of 2.53 * $10^-318$. Since the p-value is less than the the significance level I set (Î±=0.05 ) we reject the null hypothesis that there is no correlation between day trip duration and night trip duration. The Pearson R test returned a correlation coefficient of 0.998. The correlation between day trip duration and night trip duration is highly positively correlated. This indicates that as day trip duration increases, night trip duration increased. The correlation is very close to 1, which would indicate a positive linear relationship between day trip duration and night trip duration. 

## Spearman's test for correlation

H0: there is no correlation between day trip duration and night trip duration

Signficance level:  Î±=0.05

```python
spearmanr_duration = scipy.stats.spearmanr(day_sub, night_sub)
print(spearmanr_duration)
```

The Spearman test returned a p-value of 0.0. Since the p-value is less than the the significance level I set (Î±=0.05 ) we reject the null hypothesis that there is no correlation between day trip duration and night trip duration. The Pearson R test returned a correlation coefficient of 0.998. The correlation between day trip duration and night trip duration is highly positively correlated. This indicates that as day trip duration increases, night trip duration increased. The correlation is very close to 1, which would indicate a positive linear relationship between day trip duration and night trip duration.
