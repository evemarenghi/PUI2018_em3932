```python
# import packages to be used in the program
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import norm
```

## Normal Distribution

```python
# Set a seed for reproducibility 
np.random.seed(1064)

# Generate 100 samples of different sizes N (N > 100 and N < 2000) from the normal distribution
ns = 100

# Create an empty list to hold the sample data 
gauss_list = []

# Create an iterator
for i in range(ns):
    # Generate normal distributions with mean = 100, stdev = 10, and sample size between 10 and 2000
    sample = np.random.normal(loc=100, scale=10, size=np.random.rand() * 1990 + 10)
    gauss_list.append(sample)
```

```python
# For each sample plot the sample mean against the sample size (N)

# Create a figure 
fig = plt.figure(figsize=(10,8))

# Plot the sample means
for i in range(ns):
    x = gauss_list[i].size
    y = gauss_list[i].mean()
    plt.plot(x, y, 'bo', color='c')
    
# Plot the mean of the distribution 
plt.plot([0, 2000], [100,100], color='k')

# Add a title and axis labels 
plt.title('Sample Mean vs. Sample Size (N) for 100 Samples from the Normal Distribution', fontsize=16)
plt.xlabel('Sample Size (N)', fontsize=12)
plt.ylabel('Sample Mean', fontsize=12)

# Add a caption
txt = """Figure 1: Sample Mean against Sample Size (N) for 100 Samples from the Normal Distribution.
        As the sample size increases, the sample means converge to the population mean, which is 100. This illustrates
        the Central Limit Theorem."""
fig.text(.5, -0.05, txt, ha='center', fontsize=12)
```

```python
# Plot the distribution of the sample means as a histogram

# Create a figure 
fig = plt.figure(figsize=(10,8))

# Plot the distribution of the sample means from the normal distribution as a histogram 
plt.hist([gauss_list[i].mean() for i in range(ns)], color='c')

# Fit a gaussian to the distribution of the means

# Add a title and axis labels 
plt.title('Distribution of Sample Means from the Normal Distribution', fontsize=16)
plt.xlabel('Sample Mean', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add a caption
txt = """Figure 2: Distribution of Sample Means from the Normal Distribution.
        The distribution of the sample means is normally distributed and converges to 
        100, which is the population mean. This illustrates the Central Limit Theorem."""
fig.text(.5, -0.05, txt, ha='center', fontsize=12)
```

## Binomial distribution

## Chi-Squared distribution 

```python
# Set a seed for reproducibility 
np.random.seed(1064)

# Generate 100 samples of different sizes N (N > 100 and N < 2000) from the chi-squared distribution
ns = 100

# Create an empty list to hold the sample data 
chi_squared_list = []

# Create an iterator
for i in range(ns):
    # Generate chi-squared distributions with mean = 100 (100 degrees of freedom) and size between 10 and 2000 
    sample = np.random.chisquare(100, np.random.rand() * 1990 + 10)
    chi_squared_list.append(sample)
```

```python
# Plot the distribution of the sample means as a histogram

# Create a figure 
fig = plt.figure(figsize=(10,8))

# Plot the distribution of the sample means from the binomial distribution as a histogram 
plt.hist([chi_squared_list[i].mean() for i in range(ns)], color='c')

# Fit a gaussian to the distribution of the means

# Add a title and axis labels 
plt.title('Distribution of Sample Means from the Chi-Squared Distribution', fontsize=16)
plt.xlabel('Sample Mean', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add a caption
txt = """Figure 8: Distribution of Sample Means from the Chi-Squared Distribution.
        The distribution of the sample means is normally distributed and converges to 
        100, which is the population mean. This illustrates the Central Limit Theorem."""
fig.text(.5, -0.05, txt, ha='center', fontsize=12)
```

## Exponential Distribution

## Plot the distribution of all sample means

```python
# Create a list of sample means for each of the distributions 
normal_means = [gauss_list[i].mean() for i in range(ns)]
possion_means = [poisson_list[i].mean() for i in range(ns)]
binomial_means = [binomial_list[i].mean() for i in range(ns)]
chi_squared_means = [chi_squared_list[i].mean() for i in range(ns)]
exp_means = [exp_list[i].mean() for i in range(ns)]

# Combine all sample means into one list
all_sample_means = normal_means + possion_means + binomial_means + chi_squared_means + exp_means
```

```python
# Plot a histogram of all sample means (together for all distributions)

# Create a figure 
fig = plt.figure(figsize=(10,8))

# Plot the distribution of all sample means in a histogram 
plt.hist(all_sample_means, color='c')

# Fit a gaussian to the distribution of the means

# Add a title and axis labels 
plt.title('Distribution of All Sample Means for All Distributions', fontsize=16)
plt.xlabel('Sample Mean', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add a caption
txt = """Figure 11: Distribution of All Sample Means for All Distributions.
        The distribution of all sample means is normally distributed and converges to 100, 
        which is the population mean for each distribution. This illustrates the Central Limit Theorem."""
fig.text(.5, -0.05, txt, ha='center', fontsize=12)
```
