# Task 2 â€” Working with date/time

```python
# Selecting trips started on Feb-02-2015 only
query = '''
SELECT * 
FROM fb55.citibike
WHERE EXTRACT(DAY FROM starttime) = 2
'''

table = get_data(query)
table.head(5)
```

```python
# Selecting trips started on the weekends - What are average trip duration during weekends?
query = '''
SELECT day, AVG(tripduration) AS avg_duration
FROM
(SELECT EXTRACT(DAY FROM starttime) as day, tripduration 
FROM fb55.citibike) AS subquery
WHERE day = 1 OR day = 2
GROUP BY day
ORDER BY day
'''

table = get_data(query)
table.head(10)
```

```python
# What are average trip duration during weekdays?
query = '''
SELECT day, AVG(tripduration) AS avg_duration
FROM
(SELECT EXTRACT(DAY FROM starttime) as day, tripduration 
FROM fb55.citibike) AS subquery
WHERE day != 1 AND day != 2
GROUP BY day
ORDER BY day
'''

table = get_data(query)
table.head(10)
```

# Task 3 â€” Working with Space

```python
# Showing the list of start station locations 
query = '''
SELECT start_station_name
FROM fb55.citibike
GROUP BY start_station_name
'''

table = get_data(query)
table.head(20)
```

```python
# Showing the number of trips started per station 
query = '''
SELECT start_station_name, COUNT(*) as trips
FROM fb55.citibike
GROUP BY start_station_name
ORDER BY trips DESC
'''

table = get_data(query)
table.head(20)
```

```python
# Showing the number of trips started for stations within 500m of Time Square!
# The coordinates of Time Square is (40.7577,-73.9857)
# One degree of latitude = 111.03 km 
# 500m = 0.004503 degrees 
# One degree of longtidue = 85.39 km
# 500m = 0.00585 degrees 
query = '''
SELECT start_station_name, COUNT(*) as trips
FROM
(SELECT *
FROM fb55.citibike
WHERE start_station_latitude < (40.7577 + 0.004503)
AND start_station_latitude > (40.7577 - 0.004503)
AND start_station_longitude < (-73.9857 + 0.00585)
AND start_station_longitude > (-73.9857 - 0.00585) ) AS subquery 
GROUP BY start_station_name
ORDER BY trips DESC
'''

table = get_data(query)
table.head(5)
```

# Task 4 â€” Putting it all together

```python
# Find the station that had the longest average trip duration during
# weekends and within 500m of TimeSquare! 
query = '''
SELECT start_station_name, AVG(tripduration) as avg_duration
FROM 
(SELECT start_station_name, tripduration, EXTRACT(DAY FROM starttime) as day 
FROM fb55.citibike
WHERE start_station_latitude < (40.7577 + 0.004503)
AND start_station_latitude > (40.7577 - 0.004503)
AND start_station_longitude < (-73.9857 + 0.00585)
AND start_station_longitude > (-73.9857 - 0.00585) ) AS subquery
WHERE day = 1 OR day = 2
GROUP BY start_station_name
ORDER BY avg_duration DESC
LIMIT 1
'''

table = get_data(query)
table.head(5)
```

```python
# Extra: create lines for trips started from stations within 500m of Times
# Squares and lasted less than 2 hours. The number of trips per each
# pair of stations are output as attributes of these lines. 
query = '''
SELECT start_station_name, end_station_name, COUNT(*) as trips
FROM fb55.citibike
WHERE start_station_latitude < (40.7577 + 0.004503)
AND start_station_latitude > (40.7577 - 0.004503)
AND start_station_longitude < (-73.9857 + 0.00585)
AND start_station_longitude > (-73.9857 - 0.00585)
AND tripduration < (60 * 60 * 2)
GROUP BY 1, 2
ORDER BY trips DESC
'''

table = get_data(query)
table.head(20)
```
